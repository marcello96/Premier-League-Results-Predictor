```python
def getMean(df, column, team_column_name, team):
    team_data = mdata.loc[mdata[team_column_name] == team]
    return team_data[column].mean(axis = 0)
```

```python
stats = pd.DataFrame(columns=['Team', 'HomeGoals', 'AwayGoals', 'HomeFouls', 'AwayFouls', 'HomeYellows',
                              'AwayYellows', 'HomeReds', 'AwayReds'])

home_team_column = 'HomeTeam'
away_team_column = 'AwayTeam'

for team in mdata['HomeTeam'].unique():
    h_mean_goals = getMean(mdata, 'HomeGoals', home_team_column, team)
    h_mean_fouls = getMean(mdata, 'HomeFouls', home_team_column, team)
    h_mean_yellow = getMean(mdata, 'HomeYellow', home_team_column, team)
    h_mean_red = getMean(mdata, 'HomeRed', home_team_column, team)
    a_mean_goals = getMean(mdata, 'AwayGoals', away_team_column, team)
    a_mean_fouls = getMean(mdata, 'AwayFouls', away_team_column, team)
    a_mean_yellow = getMean(mdata, 'AwayYellow', away_team_column, team)
    a_mean_red = getMean(mdata, 'AwayRed', away_team_column, team)
    
    stats = stats.append({'Team': team, 'HomeGoals': h_mean_goals, 'AwayGoals': a_mean_goals, 
                          'HomeFouls': h_mean_fouls, 'AwayFouls': a_mean_fouls, 
                          'HomeYellows': h_mean_yellow, 'AwayYellows': a_mean_yellow,
                          'HomeReds': h_mean_red, 'AwayReds': a_mean_red}, ignore_index = True)
    
stats
```

```python
def calc_card(data, referee, column_name): 
     dd = data.loc[mdata['Referee'] == referee]
     ff = dd['Away{}'.format(column_name)] + dd['Home{}'.format(column_name)]
     return ff.mean()
```

```python
def format_name(name):
    return name.lower().replace(' ', '_')

result_map = {'H': "home", 'A': "away", 'D': "draw"}
```

```python
for val in mdata['AwayTeam'].unique():
    print("team({}).".format(format_name(val)))
```

```python
for val in mdata['Referee'].unique():
    print("referee({}).".format(format_name(val)))
```

```python
for _, row in mdata.iterrows():
    print("match({}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {})."
          .format(format_name(row['HomeTeam']), format_name(row['AwayTeam']), row['HomeGoals'], row['AwayGoals'],
                  result_map[row['Result']], format_name(row['Referee']), row['HomeShots'], row['AwayShots'],
                  row['HomeShotsTarget'], row['AwayShotsTarget'], row['HomeCorners'], row['AwayCorners'],
                  row['HomeFouls'], row['AwayFouls'], row['HomeYellow'], row['AwayYellow'], row['HomeRed'], 
                  row['AwayRed']))
```

```python
r_stats = pd.DataFrame(columns=['Referee', 'Yellows', 'Reds'])

for referee in mdata['Referee'].unique():
    yellows = calc_card(mdata, referee, 'Yellow')
    reds = calc_card(mdata, referee, 'Red')
    r_stats = r_stats.append({'Referee': referee, 'Yellows': yellows, 'Reds': reds}, ignore_index = True)
    
r_stats
```
